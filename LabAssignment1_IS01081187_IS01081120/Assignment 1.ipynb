```python
# Nur Amira Hanina Binti Khairul Nazmi (IS01081120)
# Marsya Nur Iman Binti Mohd Ishak (IS01081187)

import csv #imports the csv module
import datetime #provide functions to work with dates and times
import requests #make HTTP requests to web APIs
import re #provide regular expressions to match pattern in the URL

# URL for Shopee product page
url = 'https://shopee.com.my/-Max-6-packs-Dendeng-Nyet-Berapi-by-Khairulaming-i.364740282.25363171499?sp_atk=b600ac55-0f23-4b9b-af94-18a0d7fbca5d'

# Function to find pattern in the URL
r = re.search(r'i\.(\d+)\.(\d+)', url)
if r:
    shop_id, item_id = r.group(1), r.group(2) #to extract the shop ID and item ID from capture groups
else:
    print("Invalid URL")
    exit()


def convert_ctime_to_datetime(ctime_value):
    """Converts a ctime value (potentially a timestamp) to a datetime object.

    Args:
        ctime_value: The value retrieved from the 'ctime' field in the API response.

    Returns:
        A datetime object representing the converted time, or None if conversion fails.
    """

    try:
        timestamp = int(ctime_value)
        datetime_obj = datetime.datetime.fromtimestamp(timestamp) # Create datetime object
        return datetime_obj.strftime("%Y-%m-%d %H:%M")  # Format as "YYYY-MM-DD HH:MM"
    except (ValueError, TypeError):
        print(f"Error converting ctime value: {ctime_value}")
        return None


def extract_date(datetime_str):
    """Extracts only the date portion (YYYY-MM-DD) from a formatted datetime string.

    Args:
        datetime_str: The string containing the date and time in a format like "YYYY-MM-DD HH:MM".

    Returns:
        The extracted date portion as a string (YYYY-MM-DD), or None if the format is invalid.
    """

    if datetime_str:
        try:
            # Split the string at the space separator
            date_part, _ = datetime_str.split(" ")
            return date_part
        except ValueError:
            print(f"Invalid datetime format: {datetime_str}")
            return None
    else:
        return None


offset = 0
with open('Shopee_Reviews.csv', 'w', newline='') as csvfile:  # Open CSV file in write mode
    fieldnames = ['Review Date', 'Author Name', 'Comment']  # Define CSV header
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)  # Create CSV writer object
    writer.writeheader()  # Write header row

    while True:
        ratings_url = f'https://shopee.com.my/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0'

        try:
            # Send API request and handle potential errors
            response = requests.get(ratings_url)

            data = response.json()

            # Limit for the number of pages (3 pages)
            page_limit = 3

            if 'data' in data and 'ratings' in data['data'] and data['data']['ratings']:
                # To check if the API response contains the expected structure with ratings data
                for rating in data['data']['ratings']: 
                    try:
                        datetime_str = convert_ctime_to_datetime(rating['ctime'])
                        date_only = extract_date(datetime_str)

                        if date_only:
                            writer.writerow({
                                'Review Date': date_only,
                                'Author Name': rating['author_username'],
                                'Comment': rating['comment']
                            })
                        else:
                            print("Failed to extract date from ctime")
                    except KeyError:
                        print("Review is missing a 'ctime' field")
                    print(rating['author_username'])
                    print(rating['comment'])
                    print('-' * 80)  # Separator between reviews 
            else:
                print("No ratings found in the response")

        except requests.exceptions.RequestException as e:
            print(f"Error making API request: {e}")

        if offset >= (page_limit - 1) * 20:  # Check if offset reaches the limit
            break 
        
        offset += 20
```
